```python
import cv2
import SimpleITK as sitk
import pydicom
```

```python
input = 'C://Users//Chuan//Desktop//Ren//SongZhengci//1.dcm'
output = 'C://Users//Chuan//Desktop//Ren//SongZhengci//Output.avi'

def loadFile(input):
    ds = sitk.ReadImage(input)
    img_array = sitk.GetArrayFromImage(ds)
    frame_num, width, height = img_array.shape
    return img_array, frame_num, width, height

img_array, frame_num, width, height = loadFile(input)

fourcc = cv2.VideoWriter_fourcc('M','J','P','G') #Microspoft Video 1

frame_num, width, height = img_array.shape
    
video = cv2.VideoWriter(output, fourcc, 15, (width, height)) # Initialize Video File   

for frame in img_array:
        
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_GRAY2BGR)
    video.write(frame_rgb)
    print(frame_rgb.shape)
    print(frame_rgb)
    print(frame.shape)
    print(frame)


    cv2.imshow('frame', frame_rgb)


video.release()
#cv2.destroyAllWindows() 
```
